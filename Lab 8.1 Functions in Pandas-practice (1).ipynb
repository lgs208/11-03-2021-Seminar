### **Task 3**. Create a new column that discretize the value of sales to low: <200, mid: >200 and <1000, high: >1000

```python
def value_filter(x):
    if x < 200:
        return 'low'



    if (x > 200) & (x < 1000):
        return 'mid'



    if x > 1000:
        return 'high'



    else:
        return 'Nan'

    df['value'].apply(value_filter)
```

```python
df['value'].apply(value_filter)
```
